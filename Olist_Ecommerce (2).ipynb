```python
seller_review_score=seller_review_score_mean.groupby("review_score")["seller_id"].count().sort_values(ascending=False)
seller_review_score=pd.DataFrame(seller_review_score)

seller_review_score.rename(columns={"seller_id":"seller_count"},inplace=True)
#seller_review_score.to_csv("seller_review_score.csv")
seller_review_score
```

```python
#visualize
plt.figure(figsize=(10,5))
sns.barplot(data=seller_review_score,
            x="review_score",
            y= "seller_count",
            order=seller_review_score.index ,
            palette="Set2")

plt.title("Number of Sellers  as per  Review Score")
plt.xlabel("Review Score")
plt.ylabel("Seller Count")
for x , y in enumerate(seller_review_score["seller_count"]):
  plt.text(x,y ,f"{y}" , ha="center", fontsize=10)

plt.show()
```

order_review_df

Insights:

- 869 sellers have the highest possible rating of 5, indicating strong customer satisfaction and trust.

- 1,636 sellers have an average rating of 4, which is the most common rating level.

- 304 sellers are rated at 3, showing a mix of positive and negative feedback.

- 155 sellers have a low rating of 1, suggesting poor performance or customer dissatisfaction.

- 126 sellers have a rating of 2, also indicating below-average service or product experience.



The majority of sellers are rated 4 or 5, suggesting that most sellers on Olist maintain good customer service and product quality.

A smaller group (around 281 sellers, with ratings 1 or 2) may need performance improvement or quality control to enhance customer satisfaction.

```python
sales_asper_rating=orderitem_orderreview_merged_df.groupby("review_score")["order_id"].count().sort_values(ascending=False)
sales_asper_rating=pd.DataFrame(sales_asper_rating)
sales_asper_rating.rename(columns={"order_id":"order_count"},inplace=True)
sales_asper_rating
```

**Insights:**

This chart displays the number of orders placed each month from September 2016 to September 2018.

- The trend begins with just 2 orders in September 2016, gradually increasing over time.

- The highest number of orders was recorded in November 2017, with a peak of 7,423 orders.

- After a drop in December 2017, the count rose again to 7,187 orders in January 2018, followed by a gradual decline over the next several months.

- By September 2018, the number of orders had dropped significantly to just 1, indicating a potential data gap or business decline.


This chart can help identify ordering trends over time and support informed, proactive business decisions — such as adjusting inventory levels or increasing staff during periods of high demand, seasonal peaks, or promotional campaigns.
